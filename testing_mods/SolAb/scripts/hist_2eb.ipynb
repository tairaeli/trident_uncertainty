```python
def what_ions(ion_list, ray_index):
    """
    determines what ions are returned in handy dandy dataframe from salsa 
    
    :ion_list: list of ions initially passed to salsa
    
    :ray: lightray index from salsa dataframe
    """
    names = []

    for i in ion_list:

        if i in df['name'].tolist():
            if i in ray_index['name'].tolist():
                names.append(i)
            else:
                pass
        else:
            pass
    
    return names

names = what_ions(ion_list = ['Ne VIII', 'Mg X', 'O VI', 'S IV', 'Si III', 'C II', 'N I'], ray_index = df[df['lightray_index'] == 4])
names
```

```python
ray = df[df['lightray_index'] == 4]

for i in names:
    ion = ray[ray['name'] == i]
    plt.hist(ion['col_dens'], label = i)
plt.legend()
```
