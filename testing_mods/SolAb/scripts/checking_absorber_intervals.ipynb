```python
#BIG BOI PROJECT CELL
import pandas as pd

def _sort_indices(master):
    """
    gets index list ready for manipulating dataset rows
    """
    for index in range(len(master)-1):
        print(index)
        
        if index+1 >= len(master):
#             print('we out of range')
            break
    
        else:
            
            if master[index] == master[index+1]:
                del master[index+1]
            
            if master[index][1] == master[index+1][0]:
                #print(master[index][1], master[index+1][0])
                master[index].append(master[index+1][1])
                del master[index+1]
                
    return master

def _check_interval(entry1, entry2, check_num, index_list, iteration1, iteration2):
    """
    checks entries in given consecutive rows -- if they're close enough to the same value, index numbers returned as a list to be added to a master list of all indices to be changed. Used in _sort_indices()
    """
#     for index in range(len(data[name])-1):
    for num in range(check_num):
        
        if entry1+num == entry2:
            #index_list.append([index, index+1])
            index_list.append([iteration1, iteration2])
        else:
            pass

def _check_absorber_int(filename, check_num, **kwargs):
    data = pd.read_csv(filename, **kwargs)
    index_list = []
    
    for index in range(len(data['lightray_index'])-1):
        
        if index == 0:
            pass
        
        else:
            
            if data['lightray_index'][index] == data['lightray_index'][index+1]:
                
                _check_interval(data['interval_start'][index], data['interval_start'][index+1], check_num, index_list, index, index+1)
                _check_interval(data['interval_start'][index], data['interval_end'][index+1], check_num, index_list, index, index+1)
                _check_interval(data['interval_end'][index], data['interval_start'][index+1], check_num, index_list, index, index+1)
                _check_interval(data['interval_end'][index], data['interval_end'][index+1], check_num, index_list, index, index+1)
                
            else:
                
                pass
    print(f'index list: {index_list}')
    index_list.sort()
    master = _sort_indices(index_list)
    
    return master
```
